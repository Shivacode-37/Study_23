#### Masked array
*A masked array is an array where some entries are marked as invalid, missing, or not to be used.
These “masked” elements are not included in computations.*
A masked array pairs:
1. the real data (an ndarray) and
2. a mask (a boolean ndarray of the same shape).

Mask semantics:
mask==True → that element is invalid / ignored.
mask==False → element is valid and used in calculations

```python
# 1. Create a masked array of shape (4, 4) with random integers and mask the elements greater than 10. Compute the sum of the unmasked elements.
import numpy as np
import numpy.ma as ma
arr = np.random.randint(1,20, size =(4,4))
masked_arr = ma.masked_greater(arr,10)
print(arr)
print(masked_arr)


```
