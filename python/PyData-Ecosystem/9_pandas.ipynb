```python
# Arithmetic and Data Alignment
import pandas as pd
from pandas import Series
import numpy as np
s1 = pd.Series([1,2,3,4,5,6.7], index = ['a','b','c','d','e','f'])
s2 = pd.Series([6.7,5,4,3,2,1], index =['f','e','d','c','b','a'])
print(s1)
print(s2)
print(s1+s2)

# In case of DataFrame alignment is performed on both rows and columns. ****IMP.***
df1 = pd.DataFrame(np.arange(16).reshape((4,4)),index = ['mohit','shalu','chikuu','rohit'], columns =list('abcd'))
df2 = pd.DataFrame(np.arange(16).reshape((4,4)),index = ['mohit','shalu','puja','Darshan'], columns =list('abdc'))
print(df1)
print(df2)
print(df1+df2) # Adding these returns a DataFrame with index and columns that are the unions of the ones in each DataFrame.
# Matching columns will add and if indexes are present unmatched will be NAN automatically.
df3 = pd.DataFrame({'a':[1,2]})
df4 =pd.DataFrame({'b':[3,4]}) # indexes are present but columns are different
print(df3+df4) 
```

```python
# Arithmetic methods with fill values
import pandas as pd
import numpy as np
df1= pd.DataFrame(np.arange(12).reshape((3,4)),columns=list('abcd'))
df2 =pd.DataFrame(np.arange(20).reshape((4,5)),columns=list('abcde'))
df2.loc[1,'b']=np.nan
print(df1)
print(df2)
print(df1+df2)
print(df1.add(df2,fill_value=0))
print(df1.rdiv(2))
```

```python
# Operations between DataFrame and Series
#Broadcasting- means Numpy can apply operations between arrays of different shapes by stretching the smaller array along with dimension where size is 1(or missing).
import numpy as np
import pandas as pd
arr = np.arange(12.).reshape((3,4))
# print(arr)
# print(arr[0])
# print(arr-arr[0])# the subtraction is performed once for each row, this referred as Broadcasting.

frame = pd.DataFrame(np.arange(12).reshape((4,3)), columns=list("bde"))
Series=frame.iloc[0]
print(frame)
print(Series) # always show in transpose form
print(frame-Series) # it will subtract row-wise 
Series2 = pd.Series(np.arange(3),index=list('bdf'))
print(frame-Series2)
# If an index value is not found in either the Dataframe's columns or the series index the objects will be reindexed to form the union.
```

```python
# Sorting and Ranking
#sort_index method

import numpy as np
import pandas as pd
a1 = pd.Series(np.arange(4),index=['a','b','c','d'])
# print(a1)
# print(a1.sort_index())

# With DataFrame, you can sort by index on either axis:
frame = pd.DataFrame(np.arange(8).reshape((2,4)),index=["three","one"],columns=list("bdca"))
print(frame)
print(frame.sort_index())
print(frame.sort_index(axis='columns',ascending=False))

# any missing values are sorted to the end of the series by default.
a2 = pd.Series([4,3,np.nan,56,6,43,53,35,3])
print(a2.sort_values())
# Missing values can be sorted to the start instead by using the na_position
print(a2.sort_values(na_position='first'))
```

```python
# Ranking assigns ranks from one through the number of valid points in an array starting from the lowest value.
# The rank method for series and DataFrame are the place to look; by default, ranks break ties by assigning each group the mean rank:
obj = pd.Series([7,2,3,4,0,4,-5])
# print(obj)
# print(obj.rank())
# print(obj.rank(method='first'))
# print(obj.rank(ascending=False))

# DataFrame can compute rank over the rows and columns
frame = pd.DataFrame({'a':[1,2,3,4],'b':[4,5,6,7],'c':[7,8,8,9]})
print(frame)
print(frame.rank())
print(frame.rank(axis='columns'))

```
