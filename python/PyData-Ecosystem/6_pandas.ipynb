#### DataFrame
A dataframe represents a rectangluar table of data and contains an ordered,named collections of columns, each of which can be a different value type(numeric,string,Boolean etc.)
The Dataframe has both a row and column index; it can be thought of as a dictionary of series all sharing the same index.\
*While a dataframe is physically two-dimensional, you can use it to represent higher dimensional data in a tabular fromat using a hierarchical indexing.*

1. There are many ways to construct a DataFrame, though one of the most common is from a dictionary of equal  length lists or Numpy arrays:

```python
import pandas as pd
from pandas import DataFrame
data = {"district": ['kangra','mandi', 'chamba','hamirpur','kangra','kangra'],
        'population': [750000, 500000, 300000, 200000, 800000, 600000],
        'year':[2001, 2001, 2001, 2001, 2011, 2011]}
df = pd.DataFrame(data)
print(df)
print(df.head()) # first five rows
print(df.tail())# last five rows
print(df.columns) # column names
#to specific sequence of columns
df = pd.DataFrame(data, columns=['year','district','population'])
print(df)
#A column in a DataFrame can be retrieved as a Series either by dict-like notation or by attribute:
print(df['district'])
print(df.year)
print(df.loc[1]) 
print(df.iloc[0])
```

```python
val = pd.Series([2, 4, 5], index =["two","four",'five'])
print(val)
```
