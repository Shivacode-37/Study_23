```python
# Dropna()
import pandas as pd
import numpy as np
data = pd.Series([1,np.nan,5,6,7,8,8,np.nan])
print(data)
print(data.dropna()) # gives non null values
print(data[data.notna()]) # same as dropna

data1 = pd.DataFrame([[1,2,3,5.6],[1,np.nan,np.nan,3],[3,4,5,np.nan],[np.nan,np.nan,np.nan,np.nan]])
print(data1)
print(data1.dropna()) # any row who have nan value will drop
print(data1.dropna(how='all')) # it will drop only rows that are all NA.
data1[4] = np.nan # 4th column added 
print(data1)
print(data1.dropna(axis='columns',how='all')) # It will drop 4th columns because that are all nan.

df = pd.DataFrame(np.random.standard_normal((7,3)))
print(df)
df.iloc[:4,1] = np.nan # last one is exclusive
# df.loc[:4,1] = np.nan # Last one is inclusive
print(df)
print(df.dropna())
print(df.dropna(thresh=2)) # If any row have 2 nan value will drop from table.

```
