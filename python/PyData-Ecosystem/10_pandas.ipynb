#### Correlation and Covariance

1. The corr method of series computes the correlation of the overlapping, non-NA, aligned-by-index values in two series.
Relatedly, cov computes the covariance.
* .corr() .cov()

2. Using DataFrame corrwith method, you can compute pair-wise correlations between a DataFrame's columns or rows with another Series or DataFrame. Passing a Series with the correlation value computed for each column.
* .corrwith()

```python
# Unique Values Counts and Membership
import pandas as pd
import numpy as np

a = pd.Series(['c','a','a','b','c','d','d','f','m','i'])
print(a)
print(a.unique())
print(a.value_counts())
print(pd.value_counts(a.to_numpy(),sort=True))
print(pd.value_counts(a.to_numpy(),sort=False))

b = a.isin(['m','i']) # true for these two 
print(b)
print(a[b])  # holding true value

```

```python
# Index.get_indexer method, which gives you an index array from an array of possibly nondistinct values into another array of distinct values.

to_match = pd.Series(['c','c','c','r','t','y','g','h'])
uni_val = pd.Series(['c','r','t'])
indices = pd.Index(uni_val).get_indexer(to_match)
print(indices)
```

```python
# Compute on multiple related columns in a DataFrame
data=pd.DataFrame({'q1':[1,3,4,3,4],
                   'q2':[2,3,1,2,3],
                   'q3':[1,5,2,4,4]})
print(data) 
print(data['q1'].value_counts().sort_index())
result=data.apply(pd.value_counts).fillna(0)
print(result)
```
