```python
# Reindexing- It means to create a new object with the values rearranged to align with the new index.
import pandas as pd
obj = pd.Series([1,2,3,4], index =['a','b','c','d'])
print(obj)
obj2 = obj.reindex(['a','b','c','d','e','f'])
print(obj2)
```

```python
obj4 = pd.Series(['blue','purple','yellow'], index =[0,2,4])
obj4  = obj4.reindex(range(5), method = 'ffill')
print(obj4)
```

```python
# With DataFrame,reindex can alter the (row) index, columns, or both. When passed only a sequence, it reindexes the rows in the result:
import pandas as pd
import numpy as np
from pandas import DataFrame
frame = pd.DataFrame(np.arange(9).reshape((3,3)), index =['a','c','d'], columns =['Ohio','Texas','California'])
print(frame)
fr2 = frame.reindex(['a','b','c','d'])
print(fr2)
states = ['ohio','texas','california']
fr3 = frame.reindex(columns = states)
print(fr3)

# Using .loc 
frame.loc[['a','c','d'], ['Ohio','California']]
print(frame)


# Drop method will return a new object with the indicated value or values deleted from an axis.
obj = pd.Series(np.arange(5.), index =['a','b','c','d','e'])
new_obj = obj.drop('c')
print(new_obj)
```
