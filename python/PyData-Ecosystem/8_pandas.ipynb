#### Indexing, Selection and filtering
While you can select data by label this way, the preferred way to select index values is with special loc operator.
.loc
The reason to prefer loc is because of the different treatment of intgers when indexing with[]. 
Regular[]- bases indexing will treat integers as labels if the index contains integers, so the behaviour differs depending on the data type of the index. 

1. .loc[] = index integer
2. .iloc[] = anything will work it believes it is integer

```python
# Indexing, selection, and filtering
import pandas as pd
obj = pd.Series([4,7,-5,3], index =['d','b','a','c'])
obj1 = pd.Series([4,7,-5,3], index =[0,1,2,3])
# print(obj.loc[0]) # This will raise a KeyError
print(obj.iloc[0]) # This will print 4
print(obj1.loc[0]) # This will print 4
print(obj1.iloc[0]) # This will print 4
print(obj.loc['b':'c']) # Slicing with labels includes the endpoint
print(obj.iloc[1:3]) # Slicing with integer positions excludes the endpoint
# Assinging values
obj['d':'b']=6
print(obj)
```

```python
# Indexing into a dataframe retrieves one or more columns either with a single value or sequence.
# In DataFrame, slicing works in row wise.
import pandas as pd
import numpy as np
from pandas import DataFrame
data = pd.DataFrame(np.arange(16).reshape((4,4)), index =['kangra','shimla','chamba','manali'], columns =['one','two','three','four'])
print(data)
# print(data['one'])
# print(data[1:3])# slicing in row-wise
# print(data<5)
# data[data<5]=0 # Assign value
# print(data)

# Selection on Dataframe with loc and iloc
print(data.loc["kangra"])
print(data.loc[['kangra','shimla']])
print(data.loc['kangra',['two','three']])
print(data.iloc[0])
print(data.iloc[3,[0,1,2,3]])
```
