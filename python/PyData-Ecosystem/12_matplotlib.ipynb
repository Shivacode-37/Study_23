```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [1,4,9,6,12,25]
# Create a line plot
print(plt.plot(x,y))
print(plt.xlabel('x-axis'))
print(plt.ylabel('y-axis'))
print(plt.title("Line Plot"))
print(plt.show())
```

```python
# Create customized line plot
x = [1,2,3,4,5,6]
y = [1,4,9,6,12,25]
print(plt.plot(x,y,color='green',linestyle='--',marker='o',linewidth=3,markersize=9))
```

```python
# Multiple Plots
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y1 = [1,4,5,6,7]
y2 = [1,2,5,6,8]

plt.figure(figsize=(6,4))
plt.subplot(2,2,1)
plt.plot(x,y1,color="red")
plt.title("First plot")
plt.show()

plt.figure(figsize=(3,4))
plt.subplot(2,1,2)
plt.plot(x,y2, color ='violet')
plt.title('Second plot')
plt.show()

plt.figure(figsize=(3,4))
plt.subplot(2,1,2)
plt.plot(y2,x, color ='Green')
plt.title('Second plot')
plt.show()
```

```python
# Bar Plot
import matplotlib.pyplot as plt
sales = ['car','bike','bus','scotty','cycle']
values = [10,30,5,35,20]

plt.bar(sales,values, color='brown')
plt.xlabel("Vehicles")
plt.ylabel('counts of vehicle')
plt.title('Sales of yearly count')
plt.show()
```

```python
# Histograms-> are used to represent the distribution of a dataset. They divide the data into bins and count the number of data points in each bin.

data =[1,1,2,2,2,3,3,3,4,5,6,6,7,8,8,8,9,9]
# create a histogram
plt.hist(data,bins=5,color='blue',edgecolor='black')
plt.show()
```

```python
# Create a scatter plot
x =[1,2,3,4,5]
y = [6,7,8,9,10]
plt.scatter(x,y,color='red', marker='x')
plt.show()
```

```python
# pie chart
labels = ['car','bike','bus','scotty','cycle']
values = [10,30,5,35,20]
colors=['yellowgreen','violet','purple','skyblue','orange']
explode =(0.1,0,0,0,0) # move out the 1st slice
plt.pie(values,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.show()
```
