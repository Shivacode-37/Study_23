```python
# Variables- are fundamental elements in programming used to store the data that can be referenced and manipulated in a program.In Python, Variables are created when you assign a value to them.

## Naming conventions
## Variables should be descriptive and must start with an '_' and contains letters, numbers and underscores
# valid varibale name
First_name = "Mohit"
Second_name = "Mehra"

# Type checking and Type conversion
age =25
print(type(age))
#Type conversion
age_str=str(age)
print(age_str)
print(type(age_str))
```

```python
# Dynamic Typing
# Python allows the type of a variable to change as the program executes
var = 22
print(var,type(var))
var = "Shiva"
print(var,type(var))
var=3.14
print(var,type(var))
```
